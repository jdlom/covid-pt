# Confirmados por Distrito

```python
distritos = %sql SELECT distrito, confirmados_distrito_mais_recente FROM import
distritos
```

```python
df = pd.DataFrame(data=distritos)

x = df[1]
y1 = df[0]
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcdefaults()
fig, ax = plt.subplots()
y_pos = np.arange(len(y1))

ax.barh(y_pos, x, xerr=y_pos, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(y1)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('NÃºmero de Casos')
ax.set_title('Casos por Distrito')

plt.show()
```
