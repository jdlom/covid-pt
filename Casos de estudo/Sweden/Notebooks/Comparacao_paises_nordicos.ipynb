```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import datetime as dt
import sys
import os
import pandas as pd
from IPython.display import Image
from qgis.core import *
from qgis.PyQt.QtGui import *
from qgis.PyQt.QtCore import *
from IPython.display import Image

import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
QgsApplication.setPrefixPath(r'C:\\OSGeo4W64\\apps\\qgis', True)
qgs = QgsApplication([], True)
qgs.initQgis()
sys.path.append(r'C:\OSGeo4W64\apps\qgis\python\plugins')
project = QgsProject.instance()
```

```python
uri='file:///C:/OSGeo4W64/bin/Epidemiologia1/Sweden/dados/owid-covid-data2.csv?delimiter=,'
infoLyr=QgsVectorLayer(uri,'','delimitedtext')
infoLyr.isValid()
```

```python
values_case_Swe = []
values_case_Fin = []
values_case_Den = []
values_case_Nor= []
datas = []

lowerTotal = 0
i = 0


for feature in infoLyr.getFeatures():
    if feature['location']=="Sweden":
        if feature['total_cases']>0:
            values_case_Swe.append(feature['total_cases'])
          
            datas.append(feature['date'])


for feature in infoLyr.getFeatures():
    if feature['location']=="Finland":
        if feature['total_cases']>0:
            values_case_Fin.append(feature['total_cases'])
          
            #datas.append(feature['date'])

            
for feature in infoLyr.getFeatures():
    if feature['location']=="Norway":
        if feature['total_cases']>0:
            values_case_Nor.append(feature['total_cases'])
          
            #datas.append(feature['date'])

 
for feature in infoLyr.getFeatures():
    if feature['location']=="Denmark":
        if feature['total_cases']>0:
            values_case_Den.append(feature['total_cases'])
          
            #datas.append(feature['date'])
    
```

```python

fig = plt.figure(figsize=[40,10]);


cov_date = pd.Series(values_case_Swe, index=datas)
cov_date1 = pd.Series(values_case_Nor, index=datas)
cov_date2 = pd.Series(values_case_Den, index=datas)
cov_date3 = pd.Series(values_case_Fin, index=datas)
plt.xlabel('Data')
plt.tick_params(axis='x', rotation=70)
plt.ylabel('Numero de Infetados')
plt.title('Comparação casos países nórdicos', fontsize=20)
plt.plot(cov_date,  '--bo', label='Sweden')
plt.plot(cov_date1, 'o-',label='Norway')
plt.plot(cov_date2, 'o-',label='Finland')
plt.plot(cov_date3, 'o-',label = 'Denmark')

for ax in fig.axes:
  for label in ax.get_xticklabels():
    label.set_ha("left")
    label.set_rotation(45)
    
sns.set_color_codes('muted')
sns.despine(left = True, bottom = True)
fig.set_figheight(17)
fig.set_figwidth(27)
plt.tight_layout()
plt.savefig('Images/Comparacao_paises_casos.jpg', bbox_inches='tight')
plt.show()

```
