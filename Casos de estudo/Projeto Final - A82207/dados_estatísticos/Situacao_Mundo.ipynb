### CASOS CONFIRMADOS

TOP 5 Países com mais casos confirmados:

```python
confirmados = %sql SELECT country_region, confirmed FROM covid_mundo ORDER BY confirmed DESC LIMIT 5

df = confirmados.DataFrame()
df.plot.bar(figsize=(16, 4), x='country_region', y='confirmed', title='Países com mais casos confirmados - TOP 5', color='#008080')
```

Países com mais casos confirmados 

```python
mais_confirmados = %sql SELECT country_region, confirmed FROM covid_mundo WHERE confirmed <= '235000' ORDER BY confirmed DESC LIMIT 50

df = mais_confirmados.DataFrame()
df.plot.bar(figsize=(16, 8), x='country_region', y='confirmed', title='Países com mais casos confirmados', color='#008080')
```

Países com menos casos:

```python
menos_confirmados = %sql SELECT country_region, confirmed FROM covid_mundo WHERE confirmed != "0" ORDER BY confirmed ASC LIMIT 50 

df = menos_confirmados.DataFrame()
df.plot.bar(figsize=(16, 8), x='country_region', y='confirmed', title='Países com menos casos confirmados', color='#008080')
```
